```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#importing libraries
```

```python
df=pd.read_csv("/content/train.csv")
df.head()
```

```python
sns.scatterplot(df,x=df["x"],y=df["y"])
#assigning x and y axis
```

```python
df.isnull().sum()
#checking for total number of null values
```

```python
df.dropna(inplace=True)
#dropped null values
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,0].values
y=df.iloc[:,1].values
#.values is used for transforming dataframes into numpy arrays
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#20 80 is the most common split we use it for protecting our model from overfitting
#we choose random state =1 because of it offers reproduceability to the model, when ever we run the code it splits the code in a random order
```

```python
x_train=x_train.reshape(-1,1)
y_train=y_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
y_test=y_test.reshape(-1,1)
#we reshaped the values between -1 to 1
```

```python
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
# we are using r-squared for evaluation of our models performance
```

```python
plt.plot(x_test,y_pred,color="red")
sns.scatterplot(df,x=df["x"],y=df["y"])
```
